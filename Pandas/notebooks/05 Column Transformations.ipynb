```python
import pandas as pd

data = pd.read_excel("dummy_company_data.xlsx")
print(data)
```

```python
data.loc[(data["Founded Year"] > 2005), "generation"] = "New Gen"
data.loc[(data["Founded Year"] < 2005), "generation"] = "old Gen"
print(data)
```

```python
data["Company Category"] = data["Company Name"] + " (" + data["Industry"].str.upper() +")"
print(data)
```

```python
data = {"Months": ["January","Feburary", "March", "April", "May"]}
a = pd.DataFrame(data)
print(a)

def extract(value):
    return value[0:3]

a["Short_months"] = a["Months"].map(extract)
print(a)
```
