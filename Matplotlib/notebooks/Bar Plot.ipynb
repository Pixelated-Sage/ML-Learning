```python
import matplotlib.pyplot as plt

y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]
# color = ["red", "green", "yellow", "blue", "purple"]
 
plt.bar(x,y , color = "gray" , edgecolor = "gold")
plt.xlabel("Parts of harry potter", fontsize = 12 )
plt.ylabel("popularity" , fontsize = 12)
plt.title("Popularity of different parts (Harry Potter)", fontsize = 16)
plt.show
```

```python
import pandas as pd

data = pd.read_csv("dummy_company_data.csv")

df = pd.DataFrame(data)

counts = df["Industry"].value_counts()

# print(data)
plt.bar( counts.index , counts.values  ,color = "gray" , edgecolor = "gold" )
plt.xlabel("Company Name")
plt.ylabel("Revenue Generated")
```
